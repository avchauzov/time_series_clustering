```python
import pandas as pd
import seaborn as sns

from pylab import rcParams
rcParams['figure.figsize'] = 25, 10
```

# weekly

```python
dataWeekly = pd.read_csv('outputWeekly/results.csv', index_col = 0)
```

```python
dataWeekly.head()
```

```python
ax = sns.boxplot(x = 'ewmParameter', y = 'coveringRate', data = dataWeekly)
ax = sns.swarmplot(x = 'ewmParameter', y = 'coveringRate', data = dataWeekly, color = 'black')
```

```python
ax = sns.boxplot(x = 'ewmParameter', y = 'averageScore', data = dataWeekly)
ax = sns.swarmplot(x = 'ewmParameter', y = 'averageScore', data = dataWeekly, color = 'black')
```

```python
ax = sns.boxplot(x = 'ewmParameter', y = 'averageSize', data = dataWeekly)
ax = sns.swarmplot(x = 'ewmParameter', y = 'averageSize', data = dataWeekly, color = 'black')
```

```python
# weekly ewm = 0.5
```

```python
ax = sns.boxplot(x = 'thresholdParameter', y = 'coveringRate', data = dataWeekly)
ax = sns.swarmplot(x = 'thresholdParameter', y = 'coveringRate', data = dataWeekly, color = 'black')
```

```python
ax = sns.boxplot(x = 'thresholdParameter', y = 'averageScore', data = dataWeekly)
ax = sns.swarmplot(x = 'thresholdParameter', y = 'averageScore', data = dataWeekly, color = 'black')
```

```python
ax = sns.boxplot(x = 'thresholdParameter', y = 'averageSize', data = dataWeekly)
ax = sns.swarmplot(x = 'thresholdParameter', y = 'averageSize', data = dataWeekly, color = 'black')
```

```python
# weekly threshold = 0.5
```
