```python
import pandas as pd
from copy import deepcopy
```

```python
data = pd.read_csv('data/data.csv', index_col=None)
```

```python
data.head(25)
```

```python
data.fillna(0, inplace=True)
```

```python
data.head(25)
```

```python
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
```

```python
data.sort_values('Date', inplace=True)
```

```python
data.head(25)
```

```python
minDate, maxDate = min(data['Date']), max(data['Date'])
```

```python
data.set_index('Date', inplace=True)
```

```python
data = data.reindex(pd.date_range(start=minDate, end=maxDate))
```

```python
data.head(25)
```

```python
data.fillna(0, inplace=True)
```

```python
dataDaily = deepcopy(data)
dataWeekly = deepcopy(data)
```

```python
dataDaily = dataDaily.resample('D', how='sum')
dataWeekly = dataWeekly.resample('W', how='sum')
```

```python
dataDaily.head(25)
```

```python
dataWeekly.head(25)
```

```python
dataDaily.to_csv('data/1.dataDaily.csv')
dataWeekly.to_csv('data/1.dataWeekly.csv')
```
