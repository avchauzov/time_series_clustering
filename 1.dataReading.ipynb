```python
data = pd.read_csv('data/data.csv', index_col=None)
```

```python
data.head(25)
```

```python
data['Date'] = pd.to_datetime(data['Date'], dayfirst = False, yearfirst = False)
```

```python
data.head(25)
```

```python
data.sort_values('Date', inplace=True)
```

```python
data.set_index('Date', inplace=True)
```

```python
minDate, maxDate = min(data.index), max(data.index)
data = data.reindex(pd.date_range(start=minDate, end=maxDate))
```
